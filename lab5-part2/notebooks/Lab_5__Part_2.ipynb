```python
latitudes = []
longitudes = []

for i in rows:
  latitudes.append(i['latitude'])
  longitudes.append(i['longitude'])
```

```python
origin = [sum(latitudes)/len(latitudes), sum(longitudes)/len(longitudes)]
```

```python
map = folium.Map(origin,zoom_start=12)

# for coord1, coord2 in zip(latitudes, longitudes):
#   folium.Marker(location=[coord1, coord2]).add_to(map)

for i in rows:
  lat_ = i['latitude']
  long_ = i['longitude']
  # text = "operator:{}, well name:{}, well status:{}, well type:{}".format(i['operator'], i['well_name'], i['well_status'], i['well_type'])
  text = """<strong> Well Name: {}</strong> <br> <strong> Operator:{} </strong> <br>
   <strong> Well status: {} </strong> <br>
    <strong> Well type: {} </strong>""".format(i['well_name'], i['operator'], i['well_name'], i['well_type'])
  folium.Marker(location=[lat_,long_],tooltip=text).add_to(map)
```
